```python
# PCA on the ABT feature set
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from plotnine import ggplot, aes, geom_point, labs, theme_minimal

pca_df = ids_1k_abt.drop(columns=['smiles'], errors='ignore').select_dtypes(include='number')
pca_df = SimpleImputer(strategy='median').fit_transform(pca_df)
pca_df = StandardScaler().fit_transform(pca_df)

pca = PCA(n_components=2, random_state=0)
pcs = pca.fit_transform(pca_df)
pca_plot_df = pd.DataFrame({'PC1': pcs[:, 0], 'PC2': pcs[:, 1]})

print('Explained variance ratio:', pca.explained_variance_ratio_)
(
    ggplot(pca_plot_df, aes('PC1', 'PC2'))
    + geom_point(alpha=0.35, size=1)
    + theme_minimal()
    + labs(title='PCA of ABT features', x='PC1', y='PC2')
)
```

```python
# How many principal components are needed to capture most of the variance?
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

x = ids_1k_abt.drop(columns=['smiles'], errors='ignore').select_dtypes(include='number')
x = SimpleImputer(strategy='median').fit_transform(x)
x = StandardScaler().fit_transform(x)

pca_full = PCA().fit(x)
cum_var = pca_full.explained_variance_ratio_.cumsum()

for cutoff in [0.9, 0.95, 0.99]:
    n = (cum_var >= cutoff).argmax() + 1
    print(f'{n} PCs explain {cutoff:.0%} of the variance')

print('Total PCs:', len(cum_var))
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.feature_selection import VarianceThreshold

# 1. Grab numeric columns and drop completely empty ones
x_full = ids_1k.select_dtypes(include='number').dropna(axis=1, how='all')

# 2. Impute missing values
x_full = SimpleImputer(strategy='median').fit_transform(x_full)

# 3. Apply MinMaxScaler to put everything between 0 and 1
x_full = MinMaxScaler().fit_transform(x_full)

# 4. Remove features with low relative variance (e.g., < 0.005)
selector = VarianceThreshold(threshold=0.005)
x_full = selector.fit_transform(x_full)
print(f"Features remaining after thresholding: {x_full.shape[1]}")

# 5. Standardize (Normalize) the remaining features for PCA
x_full = StandardScaler().fit_transform(x_full)

# Now x_full is perfectly prepped and ready for PCA!

# 5. Run full PCA
pca_full_all = PCA().fit(x_full)
cum_var_all = pca_full_all.explained_variance_ratio_.cumsum()

print(f"Total valid numerical descriptors used: {x_full.shape[1]}")
print("-" * 40)
for cutoff in [0.9, 0.95, 0.99]:
    n = (cum_var_all >= cutoff).argmax() + 1
    print(f'{n} PCs explain {cutoff:.0%} of the variance')
```

```python
import pandas as pd
import numpy as np
from plotnine import (
    ggplot, aes, geom_line, geom_point, geom_col, geom_text, 
    theme_minimal, theme, element_text, labs, scale_y_continuous
)

# 1. Create a DataFrame containing all PCs and cumulative variance
df_var = pd.DataFrame({
    'PC': np.arange(1, len(cum_var_all) + 1),
    'Cumulative_Variance': cum_var_all
})

# 2. DYNAMICALLY filter key milestone PCs for annotations
milestones = []
percentages = [0.90, 0.95, 0.99]
for cutoff in percentages:
    n = (cum_var_all >= cutoff).argmax() + 1
    milestones.append(n)

df_milestones = df_var[df_var['PC'].isin(milestones)].copy()
df_milestones['Label'] = [f"{pc} PCs\n({pct:.0%})" for pc, pct in zip(milestones, percentages)]

# 3. Plot 1: Big Picture - Continuous Line Plot for all PCs
p1 = (
    ggplot(df_var, aes(x='PC', y='Cumulative_Variance'))
    + geom_line(color='#2b5c8f', size=1.2)
    + geom_point(df_milestones, aes(x='PC', y='Cumulative_Variance'), color='darkred', size=3)
    + geom_text(
        df_milestones,
        aes(x='PC', y='Cumulative_Variance', label='Label'),
        va='bottom',
        ha='right',
        size=9,
        nudge_x=-1,
        nudge_y=0.03,
        fontweight='bold'
    )
    + scale_y_continuous(labels=lambda l: [f"{int(x*100)}%" for x in l], limits=(0, 1.1))
    + labs(
        title=f'PCA Cumulative Explained Variance - Big Picture (All {len(cum_var_all)} PCs)',
        subtitle='Continuous curve representing cumulative variance explained by all PCs',
        x='Principal Component (PC) Number',
        y='Cumulative Variance Explained'
    )
    + theme_minimal()
    + theme(
        plot_title=element_text(size=13, fontweight='bold'),
        axis_title=element_text(size=11),
        figure_size=(10, 5)
    )
)

# 4. Plot 2: Focused view closer to 0 - Bar Plot for PCs 1 to 30
df_focus = df_var[df_var['PC'] <= 30].copy()
p2 = (
    ggplot(df_focus, aes(x='PC', y='Cumulative_Variance'))
    + geom_col(fill='#2b5c8f', width=0.8, alpha=0.85)
    + scale_y_continuous(labels=lambda l: [f"{int(x*100)}%" for x in l], limits=(0, 1.05))
    + labs(
        title='PCA Cumulative Explained Variance - Focused (PCs 1 to 30)',
        subtitle='Zoomed-in look at the first 30 components on a true numeric x-axis scale',
        x='Principal Component (PC) Number',
        y='Cumulative Variance Explained'
    )
    + theme_minimal()
    + theme(
        plot_title=element_text(size=13, fontweight='bold'),
        axis_title=element_text(size=11),
        figure_size=(10, 5)
    )
)

# 5. Display both plots in the cell output
p1
```

```python
import pandas as pd

df = ids_1k.copy()

# ============================================================
# 1. Lipinski's Rule of 5 (Lipinski et al., 1997)
#    Oral bioavailability — pass if ≤1 violation
# ============================================================
lipinski_violations = (
    (df['MolWt'] > 500).astype(int)
    + (df['MolLogP'] > 5).astype(int)
    + (df['NumHDonors'] > 5).astype(int)
    + (df['NumHAcceptors'] > 10).astype(int)
)
lipinski_pass = (lipinski_violations <= 1)

# ============================================================
# 2. Veber's Rules (Veber et al., 2002)
#    Oral bioavailability — must satisfy BOTH criteria
# ============================================================
veber_pass = (df['TPSA'] <= 140) & (df['NumRotatableBonds'] <= 10)

# ============================================================
# 3. Ghose Filter (Ghose et al., 1999)
#    Drug-likeness — must satisfy ALL criteria
# ============================================================
ghose_pass = (
    (df['MolLogP'].between(-0.4, 5.6))
    & (df['MolWt'].between(160, 480))
    & (df['HeavyAtomCount'].between(20, 70))
)
# Note: Ghose also uses Molar Refractivity (MR 40–130).
# Uncomment the next line if 'MolMR' exists in your dataset:
ghose_pass = ghose_pass & (df['MolMR'].between(40, 130))

# ============================================================
# 4. Egan's Rules (Egan et al., 2000)
#    Intestinal absorption — must satisfy BOTH criteria
# ============================================================
egan_pass = (df['MolLogP'].between(-1, 5.88)) & (df['TPSA'] <= 131.6)

# ============================================================
# 5. Muegge's Rules (Muegge et al., 2001)
#    Pharmacophore-point filter — must satisfy ALL criteria
# ============================================================
muegge_pass = (
    (df['MolWt'].between(200, 600))
    & (df['MolLogP'].between(-2, 5))
    & (df['TPSA'] <= 150)
    & (df['NumHDonors'] <= 5)
    & (df['NumHAcceptors'] <= 10)
    & (df['RingCount'] <= 7)
    & (df['NumRotatableBonds'] <= 15)
)

# ============================================================
# Summary Table
# ============================================================
total = len(df)
rules = {
    'Lipinski (1997)':  lipinski_pass,
    'Veber (2002)':     veber_pass,
    'Ghose (1999)':     ghose_pass,
    'Egan (2000)':      egan_pass,
    'Muegge (2001)':    muegge_pass,
}

summary = pd.DataFrame({
    'Rule': rules.keys(),
    'Pass': [v.sum() for v in rules.values()],
    'Fail': [total - v.sum() for v in rules.values()],
    'Pass %': [f"{v.sum() / total * 100:.1f}%" for v in rules.values()],
})
print(f"Total molecules: {total:,}\n")
display(summary)

# Bonus: How many pass ALL five rules simultaneously?
all_pass = lipinski_pass & veber_pass & ghose_pass & egan_pass & muegge_pass
print(f"\nPass ALL 5 rules: {all_pass.sum():,} ({all_pass.sum() / total * 100:.1f}%)")

## Calculate the average (mean) qed
avg_qed = df['qed'].mean().round(4)
print(f"Average qed: {avg_qed}")

# Calculate the median qed
median_qed = df['qed'].median().round(4)
print(f"Median qed: {median_qed}")

```

```python
# Define tests based on rules
# 1. Lipinski
ids_1k['pass_lipinski'] = (ids_1k['NumHDonors'] <= 5) & (ids_1k['NumHAcceptors'] <= 10) & (ids_1k['MolWt'] <= 500) & (ids_1k['MolLogP'] <= 5)

# 2. Veber
ids_1k['pass_veber'] = (ids_1k['NumRotatableBonds'] <= 10) & (ids_1k['TPSA'] <= 140)

# 3. Gleeson
ids_1k['pass_gleeson'] = (ids_1k['MolWt'] < 400) & (ids_1k['MolLogP'] < 4)

# 4. Ro3 (Rule of Three)
ids_1k['pass_ro3'] = (ids_1k['MolWt'] < 300) & (ids_1k['MolLogP'] <= 3) & (ids_1k['TPSA'] <= 60) & (ids_1k['NumHDonors'] <= 3) & (ids_1k['NumHAcceptors'] <= 3)

# 5. Hughes
ids_1k['pass_hughes'] = (ids_1k['MolLogP'] < 3) & (ids_1k['TPSA'] > 75)

# 6. Ghose
# Total number of atoms can be calculated exactly from existing columns: Heavy atoms + Hydrogens (approximated by mass difference)
ids_1k['TotalAtoms'] = ids_1k['HeavyAtomCount'] + ((ids_1k['MolWt'] - ids_1k['HeavyAtomMolWt']) / 1.008).round()

ids_1k['pass_ghose'] = (ids_1k['MolWt'] >= 160) & (ids_1k['MolWt'] <= 480) & \
                       (ids_1k['MolLogP'] >= -0.4) & (ids_1k['MolLogP'] <= 5.6) & \
                       (ids_1k['MolMR'] >= 40) & (ids_1k['MolMR'] <= 130) & \
                       (ids_1k['TotalAtoms'] >= 20) & (ids_1k['TotalAtoms'] <= 70)

# Identify compounds passing all tests
ids_1k['pass_all_tests'] = ids_1k['pass_lipinski'] & ids_1k['pass_veber'] & ids_1k['pass_gleeson'] & ids_1k['pass_ro3'] & ids_1k['pass_hughes'] & ids_1k['pass_ghose']

# Identify those with low QED that pass all tests (Using QED < 0.3 as a threshold for low)
low_qed_pass_all = ids_1k[(ids_1k['pass_all_tests'] == True) & (ids_1k['qed'] < 0.3)]

print(f"Total compounds passing all tests with low QED: {len(low_qed_pass_all)}")
low_qed_pass_all[['smiles', 'qed', 'MolWt', 'MolLogP', 'TPSA', 'TotalAtoms', 'MolMR']].head(20)
```
